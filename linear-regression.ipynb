#### Single Variable Regressions: 

```python
%%R

# Population estimate: 
single1 <- lm(active_shed_licenses~population_estimate, data=merged_data_new)
summary(single1)
```

```python
%%R

# Population estimate: 
merged_data_new %>% 
  ggplot(aes(population_estimate, active_shed_licenses)) +
  geom_point() +
  geom_smooth(method = "lm") + 
  labs(title = "More Populated tracts keep more active sheds", x = "Population Estimate", y = "Active Shed Count")
```
