## 2-D visualizations (aka scatter plots)

It seems simple. It's not simple. mainly because...

**CORRELATION =/= CAUSATION**

Comparing variables to one another can be TRICKY!

Let's make some 2-D visualizations (aka scatter plots).

Be VERY careful how you phrase the headlines. Take time to think through what you want to say.




```python
%%R
install.packages('ggrepel')
```

```python
%%R

# code for scatter plot 1
# be sure to include a meaningful title and subtitle

library(ggrepel) 

# Make a scatterplot of reading scores from 2004 vs 2003 
# note: we use third grade in 2003 and fourth grade in 2004
#       to theoretically track the same(ish) students
merged_data_new %>% 
  ggplot(aes(population_estimate, Age..in.years.)) +
  geom_point() +
  geom_smooth(method = "lm") + 
  labs(title = "Third and Fourth Grade Reading Scores", x = "Population Estimate", y = "Shed Age (in years)")
```

```python
%%R
merged_data_new %>% 
  ggplot(aes(population_estimate, active_shed_licenses)) +
  geom_point() +
  geom_smooth(method = "lm") + 
  labs(title = "Third and Fourth Grade Reading Scores", x = "Population Estimate", y = "Shed Age (in years)")
```

```python
%%R
merged_data_new %>% 
  ggplot(aes(population_estimate, Sidewalk.Shed.Linear.Feet)) +
  geom_point() +
  geom_smooth(method = "lm") + 
  labs(title = "Third and Fourth Grade Reading Scores", x = "Population Estimate", y = "Shed Age (in years)")
```
